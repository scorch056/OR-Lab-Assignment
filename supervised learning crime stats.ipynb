```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
state = pd.read_csv("data_STATEWISE.csv", header=3 )
educate = pd.read_csv("data_EDUCATION.csv", header=4)
```

```python
state.head()
```

```python
educate.head()
```

```python
educate = educate.set_axis(['states', 'illiterate', 'below class 10th', 'class 10th and above but below Graduation', 'graduate', 'holding technical degree/diploma', 'post graduate', 'total'], axis='columns')
```

```python
state = state.set_axis(['states', 'belongs to state', 'belongs to other state', 'belongs to other country', 'total'], axis='columns')
```

```python
state.head()
```

```python
educate.head()
```

#State Wise Dataset

```python
plt.figure(figsize=(10,15))
plt.barh(state['states'][:-2], state['belongs to state'][:-2])
plt.xticks(rotation=90)
plt.title("Same State Prisoners")
plt.show()
```

```python
ratio_same_state_prisoners = state['belongs to state']/state['total']
```

```python
plt.figure(figsize=(10,15))
plt.barh(state['states'][:-2], ratio_same_state_prisoners[:-2])
plt.xticks(rotation=90)
plt.title("Same State Prisoners (Ratio)")
plt.show()
```

Foreign Prisioners per state

```python
plt.figure(figsize=(10, 15))
plt.barh(state['states'][:-2], state['belongs to other country'][:-2], )
plt.yticks()
plt.title("Foreign Prisoners")
plt.show()
```

```python
ratio_foreign_prisoners = state['belongs to other country']/state['total']
```

```python
plt.figure(figsize=(10,15))
plt.barh(state['states'][:-2], ratio_foreign_prisoners[:-2], )
plt.xticks(rotation=90)
plt.title("Foreign Prisoners (Ratio)")
plt.show()
```

```python
education = educate[:-2]
jails = state[:-2]
education = pd.DataFrame(education)
jails = pd.DataFrame(jails)
```

```python
jails['same state prisioners normalized'] = jails['belongs to state']/jails['total']
jails['foreigner prisoners normalized'] = jails['belongs to other country']/jails['total']
education['illiterate normalized'] = education['illiterate']/education['total']
education['post graduation normalized'] = education['post graduate']/education['total']
```
